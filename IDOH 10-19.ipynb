```python
import requests
import pandas as pd
import os

# URL for the ArcGIS server feature layer
url = 'https://gisdata.in.gov/server/rest/services/Hosted/Stats_Explorer_Data/FeatureServer/1/query'

# query parameters
params = {
    'where': "ind_id LIKE 'DEATH001' AND geog_name='Marion'",
    'outFields': '*',  # Fetch all fields
    'returnGeometry': 'false',  # We don't need geometry for Excel
    'f': 'pjson'
}

# GET request to download the data
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()

    # Ensure 'features' key exists in the response
    if 'features' not in data:
        print("No 'features' key in the response. Check the server's response.")
        print(data)  # This will help diagnose any error messages from the server
    else:
        # Check if there are any features in the JSON data
        if len(data['features']) == 0:
            print('No records found for the query.')
        else:
            # Extract the attribute data from the features
            attributes = [feature['attributes'] for feature in data['features']]

            # Convert attributes to a DataFrame
            df = pd.DataFrame(attributes)

            # Save the DataFrame to an Excel file
            download_dir = os.path.expanduser("C:/Users/khazi/Downloads/")
            output_path = os.path.join(download_dir, "DEATH001_Marion1.xlsx")
            df.to_excel(output_path, index=False, encoding='utf-8')

            print('Download Successful')
else:
    print('Request failed')
```

```python
import requests
import pandas as pd
import os

url = 'https://gisdata.in.gov/server/rest/services/Hosted/Stats_Explorer_Data/FeatureServer/1/query'

params = {
    'where': '1=1',
    'outFields': 'ind_id',
    'returnDistinctValues': 'true',
    'f': 'json'
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    values = data['features']
    
    if len(values) > 0:
        ind_ids = [value['attributes']['ind_id'] for value in values]
        print('Unique ind_id values:', ind_ids)

        # Convert list of ind_ids to DataFrame
        df = pd.DataFrame(ind_ids, columns=['ind_id'])

        # Save the DataFrame to an Excel file
        download_dir = os.path.expanduser("C:/Users/khazi/Downloads/")
        output_path = os.path.join(download_dir, "Unique_ind_ids.xlsx")
        df.to_excel(output_path, index=False, encoding='utf-8')

        print('Download Successful')
    else:
        print('No values found')
else:
    print('Request failed')
```
